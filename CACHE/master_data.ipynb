```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Master_data_prac7.csv")
df.head(10)
```

```python
cricket = df[df["sport"]=="cricket"]
cricket_name = set(cricket["name"])
# print(cricket_name)

volleyball = df[df["sport"]=="volleyball"]
volleyball_name = set(volleyball["name"])
# print(volleyball_name)

hockey = df[df["sport"]=="hockey"]
hockey_name = set(hockey["name"])
# print(hockey_name) 

only_c= cricket_name.difference(volleyball_name).difference(hockey_name)
only_c

only_v=volleyball_name.difference(cricket_name).difference(hockey_name)
only_v

only_h=hockey_name.difference(cricket_name).difference(volleyball_name)
only_h

only_one=only_c.union(only_v).union(only_h)
print(len(only_one))

newdf = pd.DataFrame(df[df["name"].isin(only_one)][["id","name","sport"]])
newdf
```
